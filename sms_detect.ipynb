Data Collection & Pre-Processing

```python
import pandas as pd
#loading the data from csv file to a pandas Dataframe
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data.isnull().sum()
```

```python
# replace null values with null strings
df = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
# printing the first 5 rows of the dataframe
df.isnull().sum()
```

```python
# checking number of rows and columns in the dataframe
df.shape
```

Label Encoding

```python
# label spam mail as 0; ham mail as 1;
df.loc[df['Category'] == 'spam', 'Category',] = 0
df.loc[df['Category'] == 'ham', 'Category',] = 1
```

spam - 0
ham - 1

```python
from re import X
# separating the data as text and labels

X = df['Message']

Y = df['Category']
```

```python
print(X)
```

```python
print(Y)
```

spliting data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

feature extraction

```python
# transforming the text data to feature vectors that can beused as input to the logistic regression
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', binary=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
df.info()
```

```python
y_train= Y_train.astype('int')
y_test= Y_test.astype('int')
```

```python
X_train
```

```python
print(X_train_features)
```

```python
print(X_test_features)
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
model.fit(X_train_features, y_train)
```

```python
prediction_train_data= model.predict(X_train_features)
accuracy_train_data=accuracy_score(y_train,prediction_train_data)
```

```python
print("Accuracy on test data :", accuracy_train_data)
```

```python
prediction_test_data= model.predict(X_test_features)
accuracy_test_data=accuracy_score(y_test,prediction_test_data)
```

```python
print("Accuracy on test data :", accuracy_test_data)
```

Building Predictive System

```python
input_user_mail = ["WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only."]
input_data_features=feature_extraction.transform(input_user_mail)
```

```python
prediction = model.predict(input_data_features)
if prediction[0] ==1 :
  print("This is not a SPAM")
else:
  print("This is a SPAM")
```
